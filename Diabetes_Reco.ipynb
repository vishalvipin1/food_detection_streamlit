```python
conditions = (df['Calories'] <= 600) & \
             (df['FatContent'] <= 200) & \
             (df['SaturatedFatContent'] <= 45) & \
             (df['CholesterolContent'] <= 100) & \
             (df['SodiumContent'] <= 900) & \
             (df['CarbohydrateContent'] <= 60) & \
             (df['FiberContent'] <= 13) & \
             (df['SugarContent'] <= 12) & \
             (df['ProteinContent'] <= 40)

# Apply conditions to filter out entries
filtered_df = df[conditions]

filtered_df
```

```python
# Define conditions
conditions = (df['Calories'] <= 600.0) & \
             (df['FatContent'] <= 200.0) & \
             (df['SaturatedFatContent'] <= 45.0) & \
             (df['CholesterolContent'] <= 100.0) & \
             (df['SodiumContent'] <= 900.0) & \
             (df['CarbohydrateContent'] <= 60.0) & \
             (df['FiberContent'] <= 13.0) & \
             (df['SugarContent'] <= 12.0) & \
             (df['ProteinContent'] <= 40.0)

# Create a new column and assign labels based on conditions
df['recommended'] = [0 if condition else 1 for condition in conditions]
```

```python
df=pd.read_csv('recommended.csv')
```

```python
df['recommended'] = [1 if condition else 0 for condition in conditions]

count_recommended = df['recommended'].value_counts().get(1, 0)
count_not_recommended = df['recommended'].value_counts().get(0, 0)

print("Recommended foods:", count_recommended)
print("Not recommended foods:", count_not_recommended)
```

```python
del df['foodnumber']
```

```python
pre_final = df.iloc[:,1:]
pre_final
```

```python
X = pre_final.drop(['recommended'], axis='columns')
X
```

```python
y = pre_final['recommended']
y
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_data=scaler.fit_transform(X.to_numpy())
X_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data,y,test_size=0.2,random_state=101)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV 
param_grid = {'C': [0.1,1,10,100,1000], 'gamma': [1,0.1,0.01,0.001,0.0001]}
```

```python

grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
